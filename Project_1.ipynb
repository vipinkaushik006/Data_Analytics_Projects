# Data Cleaning

The dataset contains:
- Currency symbols (₹)
- Percentage signs (%)
- Commas in numeric values

We will clean and convert these into numeric format.


```python
# Clean rating count
data["rating_count"] = data["rating_count"].str.replace(",", "")
data["rating_count"] = pd.to_numeric(data["rating_count"], errors="coerce")

# Clean price columns
data["discounted_price"] = data["discounted_price"].str.replace("₹","").str.replace(",","")
data["actual_price"] = data["actual_price"].str.replace("₹","").str.replace(",","")

data["discounted_price"] = pd.to_numeric(data["discounted_price"], errors="coerce")
data["actual_price"] = pd.to_numeric(data["actual_price"], errors="coerce")

# Clean discount %
data["discount_percentage"] = data["discount_percentage"].str.replace("%","")
data["discount_percentage"] = pd.to_numeric(data["discount_percentage"], errors="coerce")

# Convert rating
data["rating"] = pd.to_numeric(data["rating"], errors="coerce")
```

# Handling Missing Values

We replace missing numeric values with the median to avoid bias.


```python
data.fillna({
    "rating": data["rating"].median(),
    "rating_count": data["rating_count"].median(),
    "discount_percentage": data["discount_percentage"].median()
}, inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

# Removing Irrelevant Columns

Columns such as review text and links are not required for numerical analysis.


```python
df_clean = data.drop(columns=[
    "about_product",
    "review_content",
    "review_title",
    "img_link",
    "product_link",
    "user_name",
    "user_id",
    "review_id"
])
```

```python
df_clean["price_diff"] = df_clean["actual_price"] - df_clean["discounted_price"]
df_clean.head()
```

# Correlation Analysis


```python
import seaborn as sns

# Select only numeric columns
numeric_df = df_clean.select_dtypes(include=['number'])

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()

```

# Correlation Heatmap Explanation

The correlation heatmap shows the relationship between numerical variables in the dataset.
Correlation values range from -1 to +1.

- +1 → Strong positive correlation (both variables increase together)
- 0 → No correlation
- -1 → Strong negative correlation (one increases while the other decreases)

## Observations

1. **Actual Price vs Discounted Price**
   A strong positive correlation is observed, which means products with higher original prices
   tend to also have higher discounted prices.

2. **Discount Percentage vs Price Difference**
   These variables typically show a positive correlation because a higher discount percentage
   results in a larger difference between actual and discounted price.

3. **Rating vs Rating Count**
   Usually a weak or moderate correlation is seen, indicating that products with more reviews
   may slightly tend to have stable ratings, but the relationship is not very strong.

4. **Rating vs Price**
   Little to no correlation is generally observed, meaning product price does not strongly
   influence customer ratings.

## Conclusion

The heatmap helps identify relationships between numerical features.
It shows that pricing-related variables are closely connected, while customer ratings
are relatively independent of price.

This analysis is useful for understanding which factors move together and for feature
selection in future machine learning models.

# Review Sentiment Classification

In this step, we classify reviews into three categories based on rating:

- Good Review → Rating >= 4
- Neutral Review → Rating between 3 and 4
- Bad Review → Rating < 3

This helps understand overall customer satisfaction distribution.


```python
def review_sentiment(rating):
    if rating >= 4:
        return "Good"
    elif rating >= 3:
        return "Neutral"
    else:
        return "Bad"

df_clean["review_sentiment"] = df_clean["rating"].apply(review_sentiment)

df_clean["review_sentiment"].value_counts()
```

# Review Sentiment Distribution

This chart shows the proportion of good, neutral, and bad reviews

```python
df_clean["review_sentiment"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Review Sentiment Distribution")
plt.ylabel("")
plt.show()
```
